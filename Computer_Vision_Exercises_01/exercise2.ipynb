```python
# Create a function that receive an image in grayscale and returns the mean values of the pixels of 
# that image. Test your code using blobs.tif image

def greyscaleMeanValues( greyscale_image ):
    
    return greyscale_image.mean()
    
    
img = imread('imgs/blobs.tif')
print( greyscaleMeanValues( img ))
```

```python
# Create a function that receive an image in grayscale and returns
# the mean, standard deviation, minimum, and maximum pixel values of that image. Test your code using blobs.tif image

def imageIntoStats( image ):
    
    stats = []
    
    stats.append( image.mean() )
    stats.append( image.std() )
    stats.append( image.min() )
    stats.append( image.max() )
    
    return stats

img = imread('imgs/blobs.tif')
print( imageIntoStats( img ))

```
